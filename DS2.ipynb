```python
#1
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
%matplotlib inline 
```

```python
#2 
results = []
for n in range(1,10000): 
    nums = np.random.randint(low=1,high=10, size=n)
    #choose n numbers between 1 an 10 
    mean=nums.mean()
    #find the average of this number
    results.append(mean)
    #add the average to a running list
    
```

```python
#3 
df = pd.DataFrame({ 'means' : results})
print (df.head())
#the average in the beginning are all  over the place
print (df.tail())
# as n increases, the average gets closer to 5 
df.plot(title='Law of large numbers')
plt.xlabel("Number of rows in the sample")
plt.ylsbel("Average of sample")
```

```python
def normal_pdf(x, mu = 5 , sigma = 5): 
    return (1./np.sqrt(2*3.14 * sigma**2)) * 2.718**(-(x-mu)**2 / (2. * sigma**2))

x_values = np.linspace(-20,20,100)
y_values = [normal_pdf(x) for x in x_values]
plt.plot(x_values, y_values)
```
